```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Generate a sample time series (replace this with your actual time series)
signal = imu_data["Normalized_Mag"]

# Input pattern that you want to find similar patterns to
input_pattern = signal[490:760]  # Replace this with your actual input pattern

# Function to find similar patterns using a sliding window and cosine similarity
def find_similar_patterns(input_pattern, signal, window_size, min_distance):
    similar_instances = []
    last_instance_position = -1

    for i in range(len(signal) - window_size + 1):
        window = signal[i:i + window_size]

        # Check if the current instance is far enough from the last instance
        if i - last_instance_position >= min_distance:
            # Calculate cosine similarity between the window and the input pattern
            similarity_score = cosine_similarity([input_pattern], [window])[0, 0]

            # You can set a similarity threshold based on your requirements
            similarity_threshold = 0.5

            if similarity_score > similarity_threshold:
                similar_instances.append((i, similarity_score))
                last_instance_position = i + window_size - 1  # Update last instance position

    return similar_instances

# Set the window size for the sliding window (adjust as needed)
window_size = len(input_pattern)

# Set the minimum distance between similar instances (adjust as needed)
min_distance = 20

# Find similar patterns using the sliding window approach with a minimum distance constraint
similar_instances = find_similar_patterns(input_pattern, signal, window_size, min_distance)

# Plot the original time series and highlight the similar instances
plt.figure(figsize = (25,5))
plt.plot(signal, label='Original Signal')

for instance, similarity_score in similar_instances:
    plt.plot(range(instance, instance + window_size), signal[instance:instance + window_size], 'r--')

plt.title('Similar Instances with Minimum Distance Constraint')
plt.legend()
plt.show()

# Print information about similar instances
print(f'Number of similar instances: {len(similar_instances)}')
for instance, similarity_score in similar_instances:
    print(f'Instance at index {instance} with similarity score: {similarity_score:.2f}')
```

```python
similar_instances_array = np.zeros((len(similar_instances), window_size))
for idx, (start, score) in enumerate(similar_instances):
    similar_instances_array[idx, :] = signal[start:start+window_size]


for i in range(len(similar_instances)):
    plt.plot(similar_instances_array[i, :], label=f'Shot {i + 1}')
plt.title("IMU Data for All Shots Taken")

time_series_array = similar_instances_array

# Function to calculate average percent similarity using cosine similarity
def calculate_average_similarity(data_array):
    similarity_matrix = cosine_similarity(data_array)
    np.fill_diagonal(similarity_matrix, 0)  # Exclude self-similarity
    average_similarity = np.mean(similarity_matrix)
    return average_similarity

# Calculate average percent similarity across all time series
average_similarity = calculate_average_similarity(time_series_array)

# Print the average similarity
print(f'Average Similarity Across All Time Series: {average_similarity:.2f}')

plt.text(130,-3000,f'IMU Consistency Score: {average_similarity:.2f}', fontsize = 12,
         bbox = dict(facecolor = 'teal', alpha = 0.5))
```
